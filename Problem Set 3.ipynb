# Problem Set 3 Question 1

```python

"""
Occupations
Step 1. Import the necessary libraries
"""
import pandas as pd

#Step 2. Import the dataset from the address.
#Step 3. Assign it to a variable called users
"""
Here I have used read_csv from pandas library to import the dataset and since it was seperated with |, I have additionally
added the attribute sep which allows us to define the seperator.
Moreover, to see the dataset I have used the head function which takes the number attribute to show the top n rows in the datase.
"""
users = pd.read_csv("https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user", sep = "|")
users.head(10)
```

```python
#Step 4. Discover what is the mean age per occupation
"""
To find the mean age for each occupation I have used the groupby function to split the available data into group of occupation.
Moreover, I have used the mean() function to find the mean age in each occupation
"""
me = users.groupby("occupation")
me.age.mean()
```

```python
#Step 5. Discover the Male ratio per occupation and sort it from the most to the least
"""
Here I have firest calculated total of unique values and then divided by the totaal in the particular occupation.
Moreover I have presented in the percentage.
"""
freq = users.groupby("occupation")["gender"].value_counts()
ratio = freq / users.groupby('occupation').gender.count()
print((ratio*100).sort_values(ascending = False))
```

```python
#Step 6. For each occupation, calculate the minimum and maximum ages


"""
Here I have used the aggregate function of the pandas module.
"""
print(users.groupby("occupation").age.agg([min, max]))
```

```python
#Step 7. For each combination of occupation and sex, calculate the mean age
"""
Firstly, I have grouped the dataset by the occupation and gender and then calculated the mean age.
"""
print(users.groupby(['occupation', 'gender']).age.mean())
```

```python
#Step 8. For each occupation present the percentage of women and men

"""
Calculated in the dame way as the ratio is calculated.
"""
freq = users.groupby("occupation")["gender"].value_counts()
ratio = freq / users.groupby('occupation').gender.count()
print((ratio*100))
```

# Question 6

```python
"""
Create a line plot showing the number of marriages and divorces per capita in the
U.S. between 1867 and 2014. Label both lines and show the legend.
Don't forget to label your axes!

"""

import pandas as pd
import matplotlib.pyplot as plt

df1 = pd.read_csv('us-marriages-divorces-1867-2014.csv')
df1.head()
```

```python
df1 = df1[df1['Year'].apply(lambda x: x in range(1867,2014))]
df1.plot(x = 'Year', y = ['Marriages_per_1000', 'Divorces_per_1000'])
plt.xlabel('YEAR')
plt.ylabel('Per CAPITA') 
plt.title('Marriages and Divorces per capita')
plt.legend() 
plt.show()
```

# Question 7

```python
"""
Create a vertical bar chart comparing the number of marriages and divorces per
capita in the U.S. between 1900, 1950, and 2000.
Don't forget to label your axes!
"""

df2 = df1[df1['Year'].apply(lambda x: x in [1900, 1950, 2000])]
df2.head()
```

```python
df2.plot(x="Year", y = ['Marriages_per_1000', 'Divorces_per_1000'], kind="bar", rot=5, fontsize=4)
plt.xlabel('YEAR')
plt.ylabel('Per CAPITA') 
plt.title('Bar Graph Marriages and Divorces per capita')
plt.legend() 
plt.show()
```
